```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
from matplotlib.cm import get_cmap, ScalarMappable
from matplotlib.colors import Normalize
import cartopy.crs as crs
from cartopy.feature import NaturalEarthFeature
import xarray as xr
from scipy.interpolate import griddata
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
from cartopy.mpl.ticker import (LongitudeFormatter, LatitudeFormatter,
                                LatitudeLocator, LongitudeLocator)
import matplotlib.ticker as mticker
```

```python
## Open file ##
fname = "./data/SCS/SCS_chl_data.nc"
chl_data = xr.open_dataset(fname)
```

```python
chl_data.info
```

```python
states = NaturalEarthFeature(category="cultural", scale="50m",
                             facecolor="none",
                             name="admin_0_boundary_lines_land")
cart_proj = crs.PlateCarree()
```

```python
## Select a single date (relative to SCS paper, figure 1)
date = "2005-05-17"

chl = chl_data.sel(time=date).CHL.to_numpy()
```

```python
lons = chl_data.lon.to_numpy()
lats = chl_data.lat.to_numpy()

fig = plt.figure(figsize = [10,10])
ax = plt.axes(projection=cart_proj)

ax.add_feature(states, linewidth=.5, edgecolor="black")
ax.coastlines('50m', linewidth=0.8)

colormap = plt.get_cmap("winter", 20) #or any other colormap
normalize = Normalize(vmin=0.0, vmax=0.15)


plt.contourf(lons, lats, chl, 1000,
             transform=crs.PlateCarree(),
             cmap=colormap, norm=normalize)

ax.tick_params(axis = "both", direction = "in", labelsize = 12, length = 5)
ax.set_xticks(np.arange(108,115,2), crs=crs.PlateCarree())
ax.set_yticks(np.arange(10,20,2), crs=crs.PlateCarree())
ax.xaxis.set_major_formatter(LONGITUDE_FORMATTER)
ax.yaxis.set_major_formatter(LATITUDE_FORMATTER)


ax.set_ylim([9,19])
ax.set_xlim([108,114])

cbar = plt.colorbar(ScalarMappable(norm=normalize, cmap=colormap), ax=ax, shrink=.98)
cbar.set_label(label = "Chlorophyll concentration [mg m$^{-3}$]", fontsize = 15)

plt.title("Daily CMEMS data, 17-05-2005", fontsize = 15)
plt.savefig("CMEMS1day.pdf", bbox_inches = "tight")
```

## Attempt of a 8-day average of CMEMS data

```python
## Select a single date (relative to SCS paper, figure 1)
date = ["2005-05-" + str(i) for i in range(10,18)]
chl = chl_data.sel(time=date).CHL.to_numpy()
chl.shape
```

```python
chl = np.nanmean(chl, axis = 0)
chl.shape
```

```python
lons = chl_data.lon.to_numpy()
lats = chl_data.lat.to_numpy()

fig = plt.figure(figsize = [10,10])
ax = plt.axes(projection=cart_proj)


ax.add_feature(states, linewidth=.5, edgecolor="black")
ax.coastlines('50m', linewidth=0.8)


colormap = plt.get_cmap("winter", 20) #or any other colormap
normalize = Normalize(vmin=0.0, vmax=0.15)

#plt.contour(lons, lats, chl, 10, colors="black",
#            transform=crs.PlateCarree())
plt.contourf(lons, lats, chl, 1000,
             transform=crs.PlateCarree(),
             cmap=colormap, norm=normalize)

ax.set_ylim([9,19])
ax.set_xlim([108,114])


ax.tick_params(axis = "both", direction = "in", labelsize = 12, length = 5)
ax.set_xticks(np.arange(108,115,2), crs=crs.PlateCarree())
ax.set_yticks(np.arange(10,20,2), crs=crs.PlateCarree())
ax.xaxis.set_major_formatter(LONGITUDE_FORMATTER)
ax.yaxis.set_major_formatter(LATITUDE_FORMATTER)

cbar = plt.colorbar(ScalarMappable(norm=normalize, cmap=colormap), ax=ax, shrink=.98)
cbar.set_label(label = "Chlorophyll concentration [mg m$^{-3}$]", fontsize = 15)
plt.title("8-day average data from CMEMS Aqua \n 10-17/05/2005", fontsize = 15)
plt.savefig("CMEMS8day.pdf", bbox_inches = "tight")
```

# Visualizzazione cluster

```python
chl = chl_data["CHL"][:]
lons = chl_data["lon"][:]
lats = chl_data["lat"][:]
```

```python
#fig, ax = plt.subplots(figsize = [20,20], nrows = 30, ncols = 12, subplot_kw = dict(projection=crs.PlateCarree()))
```

```python
date = ["2005-05-" + str(i) for i in range(10,31)]
chl_may = chl_data.sel(time=date)
```

```python
time = chl_may["time"][:].to_numpy()
chl = chl_may["CHL"][:]
```

```python
fig, ax = plt.subplots(figsize = [30, 20], nrows = 3, ncols = 7, subplot_kw = dict(projection=crs.PlateCarree()))

ax = ax.ravel()
normalize = Normalize(vmin=0.0, vmax=0.15)

for (i, ch) in enumerate(chl):
    ax[i].contourf(lons,lats,ch, 100, cmap = "winter", norm = normalize,
                  transform = crs.PlateCarree())
    ax[i].set_title(time[i])
```
